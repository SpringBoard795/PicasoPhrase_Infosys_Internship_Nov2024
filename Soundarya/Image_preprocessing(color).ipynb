```python
img_dir='/content/drive/MyDrive/ROCo/test/radiology/images'
```

```python
import cv2
import numpy as np
import os
```

```python
image_files=[f for f in os.listdir(img_dir) if f.endswith(".jpg")]
```

```python
image_files
```

```python
img_path=os.path.join(img_dir,image_files[4])
```

```python
img_path
```

```python
img=cv2.imread(img_path)
```

```python
img
```

```python
img.shape
```

```python
target_size=(224,224)
```

```python
cv2.resize(img,target_size)
```

```python
img_processed=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
img_normalised=img/255
```

```python
img_normalised
```

```python
img_eq=cv2.equalizeHist(img_processed.squeeze())
print(f"Equalized Image(first 5 pixels): {img_eq[:5, :5]}")
img_denoised=cv2.GaussianBlur(img_eq, (5,5), 0)
print(f"Denoised Image(first 5 pixels): {img_denoised[:5, :5]}")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
img_original=cv2.imread(img_path)
plt.imshow(cv2.cvtColor(img_original,cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')
plt.subplot(1,3,2)
plt.imshow(img_processed.squeeze(),cmap='gray')
plt.title("Processed Image")
plt.axis('off')
plt.subplot(1,3,3)
plt.imshow(img_denoised,cmap='gray')
plt.title("Equalized & Denoised Image")
plt.axis('off')
plt.show()
```

```python
import tensorflow as tf
```

```python
img=np.expand_dims(img,axis=0)
```

```python
datagen=tf.keras.preprocessing.image.ImageDataGenerator(
    rotation_range=40, width_shift_range=0.2,height_shift_range=0.2, zoom_range=0.2,horizontal_flip=0.2)
it=datagen.flow(img,batch_size=1)
augmented_images=[next(it)[0] for _ in range(5)]
```

```python
augmented_images
```

```python
plt.figure(figsize=(20,5))
for i,aug_img in enumerate(augmented_images):
  plt.subplot(1,5,i+1)
  plt.imshow(aug_img)
```
