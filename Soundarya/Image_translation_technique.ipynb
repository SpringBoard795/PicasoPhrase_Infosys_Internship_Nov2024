```python
import cv2
import matplotlib.pyplot as plt
```

```python
img_dir='/content/drive/MyDrive/ROCo/test/radiology/images'
```

```python
import numpy as np
import os
```

```python
image_files=[f for f in os.listdir(img_dir) if f.endswith(".jpg")]
```

```python
image_files
```

```python
img_path=os.path.join(img_dir,image_files[4])
```

```python
img_path
```

```python
img=cv2.imread(img_path)
```

```python
img
```

```python
image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
image_rgb
```

```python
width = image_rgb.shape[1]
height = image_rgb.shape[0]
```

```python
translation_matrix = np.array([[1, 0, 100], [0, 1, 70]], dtype=np.float32)
```

```python
translated_image = cv2.warpAffine(image_rgb, translation_matrix, (width, height))
```

```python
fig, axs = plt.subplots(1, 2, figsize=(7, 4))
axs[0].imshow(image_rgb)
axs[0].set_title('Original Image')
axs[1].imshow(translated_image)
axs[1].set_title('Image Translation')
for ax in axs:
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()
plt.show()
```
